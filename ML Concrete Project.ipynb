x: Features - All columns but last

# Split The DataSet

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

# Data Processing

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

The scaler is fit on the training data and not on testing data. Since, we are training our model on rescaled Training data and the model performs well when the testing data follows same distribution. And if the scaler is fit on testing data again, this would result in testing data having a different mean and standard deviation. Resulting in loss of performance.

Standardization is the process of transforming a variable to one with a mean of 0 and a standard deviation of 1 with shifting the mean by mu, and the standard deviation by sigma. The standardized variable is called a z-score. It is equal to the original variable, minus its mean, divided by its standard deviation.

# Linear Regression Algorithm

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

lr = LinearRegression()
lasso = Lasso()
ridge = Ridge()
lr.fit(X_train, y_train)
lasso.fit(X_train, y_train)
ridge.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
y_pred_lasso = lasso.predict(X_test)
y_pred_ridge = ridge.predict(X_test)
```

Importing models as Linear Regression and Lasso Regression and Ridge Regression and Fitting models on Training data and Making predictions on Test data
The Algorithm assigns coefficients to each input feature to form a linear relation between input features and target variable, so as to minimize an objective function. 
The objective function used in this case is Mean Squared Error. 
There are three versions of Linear Regression
    - Linear Regression - No regularisation
    - Lasso Regression - L1 regularisation (Tries to push coefficients to zero)
    - Ridge Regression - L2 regularisation (Tries to keep coefficients as low as possible)

linear equation: y = mx+b as m is coffecient and b is intercept
Regularization modifies the loss function by adding a regularization term.
Regularized Loss=Original Loss+Penalty
Here,  (lambda) is the regularization strength, which controls the trade-off between fitting the data well and keeping the model parameters small.
L1 Regularization (Lasso): The penalty is the sum of the absolute values of the parameters. This can lead to a sparse model where some parameter values are exactly zero, effectively removing those features from the model.
L2 Regularization (Ridge): The penalty is the sum of the squares of the parameters. This evenly distributes the penalty among all parameters, shrinking them towards zero but not exactly zeroing any.

Too high a value can make the model too simple and fail to capture important patterns in the data (underfitting).
Too low a value might not sufficiently penalize large coefficients, leading to a model that captures too much noise from the training data (overfitting).

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

metrics = {
    "Model": ["Linear Regression", "Lasso Regression", "Ridge Regression"],
        "Training Score" : [
        lr.score(X_train , y_train),
        lasso.score(X_train, y_train),
        ridge.score(X_train , y_train)
    ],
    "Testing Score" : [
        lr.score(X_test , y_test),
        lasso.score(X_test, y_test),
        ridge.score(X_test , y_test)
    ],
    "RMSE": [
        np.sqrt(mean_squared_error(y_test, y_pred_lr)),
        np.sqrt(mean_squared_error(y_test, y_pred_lasso)),
        np.sqrt(mean_squared_error(y_test, y_pred_ridge))
    ],
    "MSE": [
        mean_squared_error(y_test, y_pred_lr),
        mean_squared_error(y_test, y_pred_lasso),
        mean_squared_error(y_test, y_pred_ridge)
    ],
    "MAE": [
        mean_absolute_error(y_test, y_pred_lr),
        mean_absolute_error(y_test, y_pred_lasso),
        mean_absolute_error(y_test, y_pred_ridge)
    ],
    "R2": [
        r2_score(y_test, y_pred_lr),
        r2_score(y_test, y_pred_lasso),
        r2_score(y_test, y_pred_ridge)
    ]
    
}

results_df = pd.DataFrame(metrics)

results_df = results_df.round(2)

print(results_df)
```

compute performance metrics evaluation regression and store it in a dictionary convert it to dataframe and round it by 2 decimal digits

```python
def plot_predictions(ax, y_pred, y_test, title):
    ax.scatter(y_pred, y_test, s=20)
    ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
    ax.set_xlabel("Predicted Data")
    ax.set_ylabel("Actual Data")
    ax.set_title(title)
```

Plotting Predictions

plot_predictions function takes axis and predicted and actual data and title of plot as argument for reusability
of function for each algorithm

```python
models = ['Linear Regression', 'Lasso Regression', 'Ridge Regression']
predictions = [y_pred_lr, y_pred_lasso, y_pred_ridge]

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, model, y_pred in zip(axes, models, predictions):
    plot_predictions(ax, y_pred, y_test, model)

fig.suptitle("Actual vs Predicted")
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# Support Vector Algorithm

```python
from sklearn import svm
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [ 0.1, 1, 10, 100],
    'gamma': [  0.001, 0.01, 0.1, 1],
    'kernel': ['rbf', 'linear'],
}
svr = GridSearchCV(SVR(), param_grid, cv=5)
svr.fit(X_train, y_train)
```

C: Regularization parameter. A lower C value means a simpler model with potentially higher bias but lower variance, while a higher C value means a more complex model with potentially lower bias but higher variance.
gamma: Kernel coefficient for 'rbf', 'poly', and 'sigmoid'. It defines how far the influence of a single training example reaches.
kernel: Specifies the kernel type to be used in the algorithm. It can be 'rbf' (Radial Basis Function), 'poly' (Polynomial), 'linear', or 'sigmoid'.
degree: Degree of the polynomial kernel function ('poly'). It is only applicable when the kernel is 'poly'.
epsilon: Epsilon in the epsilon-SVR model. It specifies a margin of tolerance where no penalty is given to errors.

```python
best_params = svr.best_params_
print(f"Best Parameters: {best_params}")

best_svr = svr.best_estimator_
y_pred_svr = best_svr.predict(X_test)
```

Importing the GridSearchCV class for performing grid search with cross-validation.
Use 5-fold cross-validation (cv=5) to evaluate each combination of parameters.
scoring='neg_mean_squared_error' is used to select the best parameters based on the lowest mean squared error.
verbose=2 provides detailed output during the fitting process.
n_jobs=-1 uses all available CPU cores for parallel processing.
Retrieve and print the best parameters.
Use the best estimator from GridSearchCV to make predictions on the test data.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

metrics = {
    "Model": ["Support Vector Regressor"],
    "Training Score" : [svr.score(X_train , y_train)],
    "Testing Score" : [svr.score(X_test , y_test)],
    "RMSE": [np.sqrt(mean_squared_error(y_test, y_pred_svr))],
    "MSE": [mean_squared_error(y_test, y_pred_svr)],
    "MAE": [mean_absolute_error(y_test, y_pred_svr)],
    "R2": [r2_score(y_test, y_pred_svr)]
}
results_df = pd.DataFrame(metrics)
results_df = results_df.round(2)
print(results_df)

plt.scatter(y_test, y_pred_rfr)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Support Vector Regressor")
plt.show()
```

# Voting Algorithm

```python
from sklearn.ensemble import VotingRegressor
v_rgs = VotingRegressor(
    estimators=[('LR', lr), ('DT', dtr), ('RF', rfr), ('SVR' , svr)])
v_rgs.fit(X_train , y_train)
y_pred_vrgs = v_rgs.predict(X_test)
```

combines multiple regression models to improve predictive performance. Instead of relying on a single model, it aggregates the predictions of several models to produce a final prediction. The idea is that by combining different models, the ensemble can capture a wider variety of patterns in the data, leading to better generalization.
Ensemble learning is the process of combining multiple models to create a stronger overall model. It is based on the principle that a group of weak learners can come together to form a strong learner.
In the context of VotingRegressor, voting refers to the process of aggregating predictions from multiple models. For regression, this is typically done by averaging the predictions.

the terms "soft" and "hard" voting are not applicable. VotingRegressor always uses a form of soft averaging by default, where the predictions from all individual regressors are averaged to produce the final prediction.
Averaging (Soft Voting) , In VotingRegressor, soft voting refers to averaging the predictions of each individual regressor.
How it works: Each regressor in the ensemble makes a prediction for each sample. The final prediction for each sample is the average of these predictions.

In VotingRegressor, predictions are combined by averaging the outputs of all the individual regressors. This method does not involve a voting mechanism as in classification tasks, but rather a straightforward averaging of predictions. This helps to reduce variance and potentially improve the overall predictive performance compared to any single regressor.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

metrics = {
    "Model": ["Voting Classifier Regressor"],
    "Training Score" : [v_rgs.score(X_train , y_train)],
    "Testing Score" : [v_rgs.score(X_test , y_test)],
    "RMSE": [np.sqrt(mean_squared_error(y_test, y_pred_vrgs))],
    "MSE": [mean_squared_error(y_test, y_pred_vrgs)],
    "MAE": [mean_absolute_error(y_test, y_pred_vrgs)],
    "R2": [r2_score(y_test, y_pred_vrgs)]
}
results_df = pd.DataFrame(metrics)
results_df = results_df.round(2)
print(results_df)

plt.scatter(y_test, y_pred_vrgs)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Voting Classifier Regressor")
plt.show()
```

compute performance metrics evaluation regression and store it in a dictionary convert it to dataframe and round it by 2 decimal digits
